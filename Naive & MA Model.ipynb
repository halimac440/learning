# NAIVE MODEL

NAIVE STASIONER

```python
import numpy as np
import matplotlib.pyplot as plt

dt = df['DAYTON_MW'].reset_index()
# Data contoh (data time series musiman)
data = dt['DAYTON_MW']
n = len(data)

# Model Naive Stasioner
def naive_stationary_forecast(data):
    predictions = []
    for i in range(1, len(data)):
        predictions.append(data[i - 1])
    return predictions

# Model Naive Trend
def naive_trend_forecast(data):
    predictions = []
    for i in range(2, len(data)):
        predictions.append(data[i - 1] + (data[i - 1] - data[i - 2]))
    return predictions

# Model Naive Musiman
def seasonal_naive_forecast(data, seasonal_period):
    predictions = []
    for i in range(seasonal_period, len(data)):
        predictions.append(data[i - seasonal_period])  # Prediksi = nilai periode musiman pada tahun sebelumnya
    return predictions

# Membuat prediksi menggunakan ketiga model Naive
naive_stationary_predictions = naive_stationary_forecast(data)
naive_trend_predictions = naive_trend_forecast(data)
seasonal_period = 24  # Periode musiman (dalam kasus ini, 24)
naive_seasonal_predictions = naive_seasonal_forecast(data, seasonal_period)

# Plot data asli dan prediksi Naive
plt.figure(figsize=(12, 6))
plt.plot(data, label='Data Asli', marker='o')
plt.plot(range(1, n), naive_stationary_predictions, label='Prediksi Naive Stasioner', marker='x')
plt.xlabel('Waktu')
plt.ylabel('Nilai')
plt.title('Prediksi dengan Model Naive Stasioner')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(data, label='Data Asli', marker='o')
plt.plot(range(2, n), naive_trend_predictions, label='Prediksi Naive Trend', marker='^')
plt.xlabel('Waktu')
plt.ylabel('Nilai')
plt.title('Prediksi dengan Model Naive Trend')
plt.legend()
plt.grid(True)
plt.show()
```

SEASONAL

```python
plt.plot(data, label='Data Asli', marker='o')
plt.plot(range(seasonal_period, n), naive_seasonal_predictions, label='Prediksi Naive Musiman', marker='s')
plt.xlabel('Waktu')
plt.ylabel('Nilai')
plt.title('Prediksi dengan Model Naive Seasonal')
plt.legend()
plt.grid(True)
plt.show()
```

# EVALUASI

# MSE

```python
# Menghitung Prediksi dengan Ketiga Model Naive
naive_stationary_predictions = naive_stationary_forecast(data)
naive_trend_predictions = naive_trend_forecast(data)
seasonal_period = 23  # Periode musiman (dalam kasus ini, 4)
naive_seasonal_predictions = naive_seasonal_forecast(data, seasonal_period)

# Menghitung Selisih antara Prediksi dan Data Asli
def calculate_squared_error(predictions, actual_data):
    squared_errors = [(predictions[i] - actual_data[i]) ** 2 for i in range(len(predictions))]
    return squared_errors

# Mengonversi data ke dalam array NumPy
data_array = data.values

# Menghitung Selisih antara Prediksi dan Data Asli
squared_errors_stationary = calculate_squared_error(naive_stationary_predictions, data_array[1:])
squared_errors_trend = calculate_squared_error(naive_trend_predictions, data_array[2:])
squared_errors_seasonal = calculate_squared_error(naive_seasonal_predictions, data_array[seasonal_period:])


# Menghitung MSE
mse_stationary = np.mean(squared_errors_stationary)
mse_trend = np.mean(squared_errors_trend)
mse_seasonal = np.mean(squared_errors_seasonal)

print(f"MSE Naive Stasioner: {mse_stationary:.2f}")
print(f"MSE Naive Trend: {mse_trend:.2f}")
print(f"MSE Naive Musiman: {mse_seasonal:.2f}")
```

# MAD

```python
# Menghitung MAD
def calculate_mad(predictions, actual_data):
    absolute_deviation = []
    mad = np.mean(np.abs(actual_data - predictions))
    absolute_deviation.append(mad)
    return absolute_deviation


# Menghitung MAD
mad_stationary = calculate_mad(naive_stationary_predictions, data[1:])
mad_trend = calculate_mad(naive_trend_predictions, data[2:])
mad_seasonal = calculate_mad(naive_seasonal_predictions, data[23:])


# Menampilkan nilai MAD
print(f"Mean Absolute Deviation (MAD): Naive Stasioner: {mad_stationary}")
print(f"Mean Absolute Deviation (MAD): Naive Trend: {mad_trend}")
print(f"Mean Absolute Deviation (MAD): Naive Musiman: {mad_seasonal}")
```

# MAPE

```python
def calculate_mape(actual, forecast):
    n = len(actual)
    
    if n != len(forecast):
        raise ValueError("Panjang actual dan forecast harus sama")
    
    absolute_errors = [abs((actual[i] - forecast[i])/actual[i]) for i in range(n)]
    sum_absolute_errors = sum(absolute_errors)
    
    mape = (sum_absolute_errors) * 100 / n
    
    return mape

# Menghitung MAPE
mape_stationary = calculate_mape(data[1:].reset_index(drop=True), naive_stationary_predictions)
mape_trend = calculate_mape(data[2:].reset_index(drop=True), naive_trend_predictions)
mape_seasonal = calculate_mape(data[23:].reset_index(drop=True), naive_seasonal_predictions)

print(f"MAPE Naive Stasioner: {mape_stationary:.2f} %")
print(f"MAPE Naive Trend: {mape_trend:.2f} %")
print(f"MAPE Naive Musiman: {mape_seasonal:.2f} %")
```
